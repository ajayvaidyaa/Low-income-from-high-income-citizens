```python
#importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#adding headers according to the data
h=['Age','Workclass','fnlwgt','Education','Education-num','Marital-status','Occupation','Relationship','Race','Sex','Capital-gain','Capital-loss','Hours-per-week','Native-country','Salary']
```

```python
#reading the file and adding the headers to the dataset
census=pd.read_csv('adult-all (1).csv',names=h)
```

```python
census.head(10)
```

```python
#checking out missing values in workclass
census.Workclass.unique()
```

```python
#data cleaning for Wholeclass coloumn and native country
#using the mask variable name to store true/false in that variable using the != operator.
mask=(census.Workclass!="?")
mask&=(census["Native-country"]!="?")
mask.head()
```

```python
census=census[mask]
```

```python
#cleaned data
#cleaning done according to workclass and native country
census.head()
```

```python
#making sure data is cleaned and there are no ?
census.Workclass.unique()
```

```python
#making sure data is cleaned and there are no ?
census["Native-country"].unique()
```

```python
#see the values in the education field 
census['Education'].unique()
```

```python
#changing the workclass into codes 
census['workclass codes']=pd.Categorical(census['Workclass']).codes
census.head()
```

```python
#native country data into codes
census['Native Country codes']=pd.Categorical(census['Native-country']).codes
census.head()
```

```python
#native country change it to codes 
#Similar to native country do the same for salary to it codes and assign (<50K=0, >50K=1) and storing them in separte coloumn Target
census['Target']=pd.Categorical(census['Salary']).codes
census.head()
```

```python
#Assign codes to Sex and storing them in separte coloumn Gender Codes (Male=1, Female=0)
census['Gendercodes']=pd.Categorical(census['Sex']).codes
census.head()
```

```python
#Assign codes to Education and storing them in separte coloumn Education Codes
census['Education Codes']=pd.Categorical(census['Education']).codes
census.head()
```

```python
census_labels = census.Target #what we are predicting
census_labels=np.array(census_labels)
```

```python
#Here I am choosing Age, EducationNum, Gendercodes, Native country codes and workclass codes
census_data=census[['Age','Education-num','Gendercodes','Native Country codes','workclass codes']]
census_data=np.array(census_data)
```

```python
#training and testing 
trainset_data, testset_data, trainset_labels, testset_labels = train_test_split(census_data,census_labels, test_size=0.33, random_state=42)
```

```python
#prediction modeling where nearest neighbors model where K=2
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(trainset_data, trainset_labels)
```

```python
#accuracy model
neigh.score(testset_data,testset_labels)
```

```python
#prediction modeling where nearest neighbors model where K=5
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(trainset_data, trainset_labels)
```

```python
neigh.score(testset_data,testset_labels)
```

```python
#prediction modeling where nearest neighbors model where K=10
neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(trainset_data, trainset_labels)
```

```python
neigh.score(testset_data,testset_labels)
```
